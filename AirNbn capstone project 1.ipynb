```python
#To Identify the top neighborhoods_group with the most listings.
top_neighborhoods = df['neighbourhood_group'].value_counts().head(20)
print('Top Neighborhoods by Listing Count:')
print(top_neighborhoods)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_neighborhoods.index, y=top_neighborhoods.values)
plt.xlabel('Neighborhood')
plt.ylabel('Listing Count')
plt.title('Top Neighborhoods by Listing Count')
plt.xticks(rotation=45)
plt.show()
```

```python


#Most Common Room Type
most_common_room_type = df['room_type'].mode().iloc[0]
print(f'Most Common Room Type: {most_common_room_type}')
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='room_type')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.title('Room Type Distribution')
plt.xticks(rotation=45)
plt.show()
```

This bar chart displays the distribution of different room types (e.g., entire home, private room, shared room). It helps you see which type of room is most common among the listings.

```python
#Compute the correlation between latitude, longitude, and price using the corr() function.
correlation = df[['latitude', 'longitude', 'price']].corr()
print(correlation)
```

```python
#Use value_counts() to count listings in each neighborhood and select the top ones.
top_neighborhoods = df['neighbourhood'].value_counts().head(10)
print('Top Neighborhoods by Listing Count:')
print(top_neighborhoods)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_neighborhoods.index, y=top_neighborhoods.values)
plt.xlabel('Neighborhood')
plt.ylabel('Listing Count')
plt.title('Top Neighborhoods by Listing Count')
plt.xticks(rotation=45)
plt.show()
```

This bar chart highlights the neighborhoods with the highest number of listings. It allows you to identify popular areas for Airbnb rentals.


```python
#How has the availability of listings changed over time
#Convert the last_review column to datetime?
# Convert 'last_review' to datetime
df['last_review'] = pd.to_datetime(df['last_review'])
```

```python
# Group by year and month
availability_over_time = df.groupby(df['last_review'].dt.to_period("M"))['availability_365'].mean()
print('Availability Over Time:')
print(availability_over_time)
```

```python
plt.figure(figsize=(12, 6))
availability_over_time.plot(kind='line', marker='o')
plt.xlabel('Time (Year-Month)')
plt.ylabel('Average Availability (365 days)')
plt.title('Availability Over Time')
plt.grid(True)
plt.show()
```

This line chart shows how the average availability of listings changes over time, likely in months or years. It helps identify trends in property availability.

```python
#To Identify the top hosts with the most listings.
top_hosts = df['host_name'].value_counts().head(10)
print('Top Hosts by Listing Count:')
print(top_hosts)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_hosts.index, y=top_hosts.values)
plt.xlabel('Host Name')
plt.ylabel('Listing Count')
plt.title('Top Hosts by Listing Count')
plt.xticks(rotation=45)
plt.show()
```

This bar chart lists the top hosts with the most listings. It can help you identify hosts who manage a significant number of properties.
